```python
%load_ext autoreload
%autoreload 1
%aimport Data_Cleaning_Taxi
import importlib
importlib.reload(Data_Cleaning_Taxi)

import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np

import Ipynb_importer
# import Data_Cleaning_Taxi

def read(file):
    print(file)
    df = pd.read_parquet(file)
    
    print(df.head())

    if 'green' in file:
        prefix='lpep_'
    elif 'yellow' in file:
        prefix='tpep_'
        
    df[prefix+'pickup_datetime'] = pd.to_datetime(df[prefix+'pickup_datetime'], errors='coerce')
    df['pickup_date'] = df[prefix+'pickup_datetime'].dt.date
    df[prefix+'dropoff_datetime'] = pd.to_datetime(df[prefix+'dropoff_datetime'], errors='coerce')
    df['dropoff_date'] = df[prefix+'dropoff_datetime'].dt.date
    
    df = Data_Cleaning_Taxi.clean(df,prefix)
    
    df['Day of week'] = df[prefix+'pickup_datetime'].dt.dayofweek
    dt_wkends = df[df['Day of week']>4]['pickup_date'].unique()
    
    return df,dt_wkends.tolist()


def combine_df(idir,files):
    df = pd.DataFrame()
    dt_non_work = []#non-working day
    for fi in files:
        dfi,dt_wkends = read(idir+fi)
        df = df.append(dfi,ignore_index=True)
        dt_non_work+=dt_wkends
    holiday = pd.to_datetime(pd.Series(['2021-07-04','2021-07-05','2021-09-06','2021-09-16','2021-10-11'])).dt.date.unique()
    dt_non_work = np.append(dt_non_work,holiday)#Labor Day

    return df,dt_non_work


def plt_by_day(idir,title,sta_df_nwk,sta_df,key):
        
    fig = plt.figure(figsize=(20, 4), dpi=150)
    # ida_window = pd.to_datetime(['2021-08-26','2021-09-04'])
    ida_nyc_window = pd.to_datetime(['2021-09-01','2021-09-02'])
    henri_nyc_wd = pd.to_datetime(['2021-08-22','2021-08-23'])
    elsa_nyc_window = pd.to_datetime(['2021-07-08','2021-07-09'])
    noreaster_nyc_window = pd.to_datetime(['2021-10-25','2021-10-26'])
    maxy = max(sta_df[key])*1.2
    plt.fill_between(henri_nyc_wd,0,maxy, facecolor='lightskyblue', alpha=0.4)#,label='Henri_period_nyc')
    plt.fill_between(elsa_nyc_window,0,maxy, facecolor='lightskyblue', alpha=0.4)#,label='Elsa_period_nyc')
    plt.fill_between(ida_nyc_window,0,maxy, facecolor='lightskyblue', alpha=0.4)#,label='Ida_period_nyc')
    plt.fill_between(noreaster_nyc_window,0,maxy, facecolor='lightskyblue', alpha=0.4)#,label="nor'easter_period_nyc")
    plt.text(henri_nyc_wd[0],0.85*maxy,'Henri')
    plt.text(elsa_nyc_window[0],0.85*maxy,'Elsa')
    plt.text(ida_nyc_window[0],0.85*maxy,'Ida')
    plt.text(noreaster_nyc_window[0],0.85*maxy,"Nor'easter")
    # plt.fill_between(ida_nyc_window,0,maxy, facecolor='steelblue', alpha=0.4,label='Ida_period_nyc')
    
    if key=='cnt_pk':
        plt.bar(sta_df.index,sta_df['cnt_dp'],color='silver',label='cnt_dropoff_workday')
        plt.bar(sta_df_nwk.index,sta_df_nwk['cnt_dp'],color='gray',label='cnt_dropoff_non-workday')
    
    plt.plot(sta_df.index,sta_df[key],marker = 'o',color='orange',label='cnt_pickup_workday',alpha=0.5)
    plt.scatter(sta_df_nwk.index,sta_df_nwk[key],marker = 'o',c='red',label='cnt_pickup_non-workday')

    plt.xticks( sta_df.index, rotation=90 )
    plt.legend(ncol=2,loc='upper left')
    # plt.tight_layout()
    plt.margins(x=0.01, y=0.1)
    plt.title(title)
    plt.ylim((0,maxy))
    
    plt.savefig(idir+'result/'+title.replace(':','_')+'.png',format='png',bbox_inches='tight')
    plt.show()
    
    
def trips_by_day(idir,files,title):
    
    df,dt_non_work = combine_df(idir,files)
    df['index'] = df.index
    
    sta_pk = df.fillna(0).groupby(df["pickup_date"]).count()
    sta_dp = df.fillna(0).groupby(df["dropoff_date"]).count()

    sta_pd = pd.DataFrame()
    sta_pd['cnt_pk'] = sta_pk['index']
    sta_pd['cnt_dp'] =  sta_dp['index']
    print(sta_pd['cnt_pk'])
    
    sta_pd_nwk = sta_pd[sta_pd.index.isin(dt_non_work)]
    
    plt_by_day(idir,title,sta_pd_nwk,sta_pd,'cnt_pk')
    
    
def pattern_by_day(idir,files,title,key,opt,df,dt_non_work):
    
    if key=='':
        trips_by_day(idir,files,title)
        return
    
    
    if not "'s" in title:
        df[[key]].hist(log=True)
        if 'Green' in title:
            plt.savefig(idir+'result/'+'PDF_'+key+'_Green.png',dpi=100,format='png',bbox_inches='tight')
        if 'Yellow' in title:
            plt.savefig(idir+'result/'+'PDF_'+key+'_Yellow.png',dpi=100,format='png',bbox_inches='tight')
        plt.show()  
    
    
    sta_df = pd.DataFrame() 

    if opt=='sum':
        sta_df = pd.DataFrame(df.fillna(0).groupby(df["pickup_date"])[key].sum())
    elif opt=='mean':
        sta_df = pd.DataFrame(df.fillna(0).groupby(df["pickup_date"])[key].mean())
    elif opt=='median':
        sta_df = pd.DataFrame(df.fillna(0).groupby(df["pickup_date"])[key].median())

    sta_df_nwk = sta_df[sta_df.index.isin(dt_non_work)]
    
    plt_by_day(idir,title,sta_df_nwk,sta_df,key)
    


    
    
def run(idir,files,_title):
        
    keys_dict = {'Trips':['',''],
                 'Ridership':['passenger_count','sum'],
                 'Trip Distances':['trip_distance','sum'],
                 'Average Trip Distances':['trip_distance','mean'],
                 'Median Trip Duration':['delta_time','median'],
                 'Median Trip Speed':['speed','median'],
                 'Average Tips':['tip_amount','mean']
                 # 'Stationarity of Zone trips':['',''],
                 
                }
    df,dt_non_work = combine_df(idir,files)
    
        
    for k,v in keys_dict.items():
#         if "PickupID" in k:
#             # print(df.groupby(["PULocationID",'pickup_date']).count())
#             # tmp = pd.DataFrame(df.groupby(["PULocationID",'pickup_date'])[v[0]].sum()).reset_index()
#             gps = df.groupby(["PULocationID"])
#             for nm,df2 in gps:
#                 nm = str(nm)
#                 title = _title + 'Zone '+nm +"'s Pickups"+' by Day'
#                 pattern_by_day(idir,files,title,v[0],v[1],df2,dt_non_work) 
#                 return
       
#         else:        
            title = _title + k +' by Day'
            pattern_by_day(idir,files,title,v[0],v[1],df,dt_non_work) 


    


if __name__=='__main__':
    
    input_dir = "D:\【Columbia】\project\data\TLC/"
    items = [i for i in os.listdir(input_dir) if os.path.splitext(i)[1] == '.parquet']
    G_fi = [i for i in items if 'green' in i]
    Y_fi = [i for i in items if 'yellow' in i]
    _G_title = 'Temporal Pattern of Green Taxi: '
    _Y_title = 'Temporal Pattern of Yellow Taxi: '
    
    run(input_dir, G_fi, _G_title)
    run(input_dir, Y_fi, _Y_title)
    
```
